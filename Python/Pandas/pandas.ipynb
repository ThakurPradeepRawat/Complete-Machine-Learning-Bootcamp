# Pandas - DataFrame And Series 

Pandas is  a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures : Series and DataFrame . A Series is a one - dimensional array-like object, while a DataFrame is a two - dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns)

```python
import pandas as pd 
```

## Series 

it is a one dimensional labelled array capable of string / holding homogeneous data of any type(integers , float , string , _ _ _ _ _ , Python objects etc )

```python
 # creatng a series using list 
lst = [10,20,30,40,50,60]
series = pd.Series(lst)
print("series :- ")
print(series)
print("Type of series :- " , type(series))
```

```python
# creating a series using dictionary 
dct = {'Name ' : "Pradeep" , 'Branch ' : " AI-ML" , 'Course': "B-Tech" }
series_dict = pd.Series(dct)
print("Series :- ")
print(series_dict)
print("Type of series :- " , type(series_dict))
```

```python
# creating an series object with programmer - defined index 
lst = [10,"Rossum", 34.56,"Author"]
series = pd.Series(lst , index=['StNo' , "Name ","Marks" , "Desg"])
print(series)
```

```python
#get first Series values 
print('First Name :- ',series[0])
```

```python
# update first value of series 
series[0]=20
series
```

```python
lst = [10,20,30,40,50,60]
s = pd.Series(lst)
print("*"*50)
print(s)
print("*"*30)
print("product of values : - " ,s.product())
print("Cummulative product of values :- \n " , s.cumprod())
print("*"*50)
print("Maximum value  of series :- " , s.max())
print("Minimum value of series  :- " , s.min())
print("Mean of Series :- " , s.mean())
print("Median of series :- " , s.median())
print("standard deviatation of series :- " , s.std())
print("Sum of values of series :- " , s.sum())
print("*"*30)
print("Sort the series in descending order :- \n" , s.sort_values(ascending = False))
print("*"*30)
```

```python
print("*"*30)
print("Total count of values in series :- " , s.count())
print("Total null or Nan values in series :- " , s.isnull().sum())
print("Unique values in series :- " , s.unique())
print("Total unique values in series :- " , s.nunique())
print("*"*30)
```

```python
# bottom 5 values 
s.tail()
```

```python
# statistical description of series 
s.describe()
```

# DataFrame

DataFrame  is one of the class and whose object allows us to store the data in tabular format or 2 - dimensional format  and where we  can perform data analysis.

## Attributes of DataFrame 

There are 10 attributes in the DataFrame
                     
                      
                                1. index
                                2. columns
                                3. axes
                                4. dtypes 
                                5. size
                                6. shape
                                7. ndim 
                                8. empty
                                9. T (Transpose)
                                10. Values 

```python
lst_dict = {
    "name": ["Pradeep Rawat" , "Ashish Lodhi " , "Shashkant Sahu " , " Aman "],
    "Marks " : [100,50,23,42],
    "Age" : [21,20,18,13]
}
df = pd.DataFrame(lst_dict)
df
```

```python
print("Range of index :- ",df.index)
print("Column names :-" , df.columns)
print("Range and columns :-" , df.axes)
```

```python
print("Data types of columns :- \n " , df.dtypes)
```

```python
print("Total size of df :- " , df.size)
```

```python
print("Shape of DataFrame :- " , df.shape)
```

```python
print("Transpose of DataFrame :- \n")
df.T
```
