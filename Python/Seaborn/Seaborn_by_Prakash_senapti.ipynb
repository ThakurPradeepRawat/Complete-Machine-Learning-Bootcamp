```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,6))
```

```python
sns.scatterplot(x="total_bill", y="tip", data = tips, hue='time', size='size',palette='deep')
plt.title("Scatterplot of Total Bill Vs Tip ")
plt.show()
```

```python
## Line Plot 
sns.lineplot(data = tips , x="size" , y='total_bill', hue="sex", markers="o")
plt.title("Lineplot of Total Bill Vs Size")
plt.show()
```

```python
## Line Plot 
sns.lineplot(data = tips , x="size" , y='total_bill', hue="sex", markers="o" , ci=None)
plt.title("Lineplot of Total Bill Vs Size")
plt.show()
```

## Barplot 

```python
sns.barplot(data=tips, x='day', y="total_bill", hue='sex' , palette="muted")
plt.title("Barplot of Total bill vs day")
plt.show()
```

```python
## boxplot 
```

```python
tips.columns
```

```python
sns.boxplot(data=tips,x='day',y='tip' , hue='smoker', palette="Set2")
plt.title("Boxplot of days vs tip ")
plt.show()
```

```python
sns.violinplot(data=tips, x='day',y='total_bill', hue='time', palette="pastel", split=True)
plt.title("Violinplot of days vs total_bill ")
plt.show()
```

```python
## countplot
```

```python
sns.countplot(data=tips, x='day',hue='smoker', palette="dark")
plt.title("Countplot of days vs smoker ")   
plt.show()
```

```python
tips.columns
```

```python
# regression plot
sns.regplot(data=tips,x='total_bill', y='tip',scatter_kws={'color':'red'}, line_kws={'color':'blue'})
plt.title("Regression plot of total_bill vs tip")
plt.show()
```

```python
## Histogram
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='blue', alpha=0.5)   
plt.title("Histogram of total_bill")    
plt.show()
```

```python
# 9 pairplot
sns.pairplot( tips, vars=["total_bill", "tip", "size"], hue="time", palette="husl")
plt.title("Pairplot of total_bill, tip and size")
plt.show()       
```

```python
## catplot
sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='point', palette="bright")
plt.title("Catplot of days vs tip")
plt.show()
```

```python
## jointplot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', color='purple' , hue='smoker')
plt.title("Jointplot of total_bill vs tip")
plt.show()
```

```python
## FacetGrid
sns.FacetGrid(tips,col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill', kde=True)
plt.subplots_adjust(top=0.9)  # Adjust the top margin to make room for the title
plt.suptitle("FacetGrid of total_bill by time and smoker status")
plt.show()
```

```python
## strripe plot
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', jitter=True, palette="Set1")
plt.title("Stripplot of days vs total_bill")
plt.show()
```

```python
## kde plot
sns.kdeplot(data=tips,x='total_bill', hue = 'sex',palette='tab10' , fill=True, alpha=0.5)
plt.title("KDE plot of total_bill vs gender")
plt.show()
```
